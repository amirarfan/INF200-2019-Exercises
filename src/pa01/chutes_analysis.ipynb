## Simulation of games with 1, 2, 4 and 8 players

```python
from snakes_and_ladders import multi_game_experiment
from matplotlib import pyplot as plt
```

```python
game_durations = []
players = [1,2,4,8]

for player in players:
    number_of_moves, duration = multi_game_experiment(100,player,2)
    game_durations.append(duration)

plt.hist(game_durations[0], color='b', label='1 Player', histtype='step')
plt.hist(game_durations[1], color='g', label='2 Players', histtype='step')
plt.hist(game_durations[2], color='r', label='4 Players', histtype='step')
plt.hist(game_durations[3], color='y', label='8 Players', histtype='step')
plt.legend()
plt.title('Durations of games with different amount of players')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.figure(figsize=(20,10))
plt.show()
```
